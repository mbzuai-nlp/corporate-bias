```python
import os
import inspect
from dotenv import load_dotenv
from openrouter import OpenRouter

load_dotenv()

GEMINI_API_KEY = os.environ["GEMINI_API_KEY"]

# Gemini's OpenAI-compatible base URL.
# The OpenRouter SDK calls this `server_url`.
GEMINI_OPENAI_SERVER_URLS = [
    "https://generativelanguage.googleapis.com/v1beta/openai",
    "https://generativelanguage.googleapis.com/v1beta/openai/",
]

print("OpenRouter constructor:")
print(inspect.signature(OpenRouter))

last_error = None

for server_url in GEMINI_OPENAI_SERVER_URLS:
    print(f"\nTrying server_url={server_url!r}")

    try:
        client = OpenRouter(
            api_key=GEMINI_API_KEY,
            server_url=server_url,
            timeout_ms=30000,
        )

        response = client.chat.send(
            model="gemini-2.5-flash",
            messages=[
                {
                    "role": "user",
                    "content": "Say exactly: hello from direct gemini",
                }
            ],
            timeout_ms=30000,
            reasoning_effort="low"
        )

        print("\nSUCCESS")
        print("\nRAW RESPONSE:")
        print(response)

        print("\nTEXT:")
        print(response.choices[0].message.content)

        break

    except Exception as e:
        last_error = e
        print("FAILED:", type(e).__name__, str(e))

else:
    raise RuntimeError(
        "OpenRouter SDK constructed with server_url, but the request still failed. "
        "Either the SDK generated paths/auth do not match Gemini's OpenAI-compatible API, "
        "or the Gemini endpoint rejected the request."
    ) from last_error
```

```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()

GEMINI_API_KEY = os.environ["GEMINI_API_KEY"]

client = OpenAI(
    api_key=GEMINI_API_KEY,
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/",
    timeout=30.0,
)

response = client.chat.completions.create(
    model="gemini-2.5-flash",
    reasoning_effort="low",  # use "none" to disable thinking on Gemini 2.5 Flash
    messages=[
        {
            "role": "user",
            "content": "Say exactly: hello from direct gemini",
        }
    ],
)

print("\nRAW RESPONSE:")
print(response)

print("\nTEXT:")
print(response.choices[0].message.content)
```
