```python
import os
import inspect
from openrouter import OpenRouter

GEMINI_API_KEY = os.environ["GEMINI_API_KEY"]
GEMINI_OPENAI_BASE_URL = "https://generativelanguage.googleapis.com/v1beta/openai/"

print("OpenRouter constructor:")
print(inspect.signature(OpenRouter))

# Try common base-url kwarg names used by SDKs.
client = None
last_error = None

for kwargs in [
    {"api_key": GEMINI_API_KEY, "base_url": GEMINI_OPENAI_BASE_URL},
    {"api_key": GEMINI_API_KEY, "base_url": GEMINI_OPENAI_BASE_URL.rstrip("/")},
    {"api_key": GEMINI_API_KEY, "api_base": GEMINI_OPENAI_BASE_URL},
    {"api_key": GEMINI_API_KEY, "base_url_override": GEMINI_OPENAI_BASE_URL},
]:
    try:
        print("\nTrying:", kwargs.keys())
        client = OpenRouter(**kwargs)
        print("Constructed client:", client)
        break
    except Exception as e:
        last_error = e
        print("Failed:", type(e).__name__, str(e))

if client is None:
    raise RuntimeError(
        "Could not construct OpenRouter client with a Gemini base URL. "
        "The OpenRouter SDK likely does not expose base_url override."
    ) from last_error

response = client.chat.send(
    model="gemini-2.5-flash",
    messages=[
        {"role": "user", "content": "Say exactly: hello from direct gemini"}
    ],
    timeout_ms=30000,
)

print("\nRAW RESPONSE:")
print(response)

print("\nTEXT:")
print(response.choices[0].message.content)
```
